```python
import torch
from tokenizer import Tokenizer
```

```python
t = Tokenizer()
t.fit("the cat sat on the mat")
```

```python
embedding = torch.nn.Embedding(num_embeddings=len(t.words_2_idx), embedding_dim=16)
```

```python
encoding = t.encode("the cat! sat haha on the mat haha")
```

```python
encoding_tensor = torch.Tensor(encoding).int()
```

```python
output = embedding(encoding_tensor)
```
