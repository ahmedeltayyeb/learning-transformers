# Positinal Encoding

```python
import torch
import math
```

```python
def positional_encoding(position, d_model):
    pe = torch.zeros(d_model)
    for i in range(d_model):
        if i % 2 == 0:
            pe[i] = math.sin(position / (10000 ** (2 * (i // 2) / d_model)))
        else:
            pe[i] = math.cos(position / (10000 ** ((2 * (i // 2)) / d_model)))
    return pe
```

```python
sentence = "the cat sat on the mat"
d_model = 4

for i, word in enumerate(sentence.split()):
    print(f"Word: {word}, Position: {i}, Positional Encoding: {positional_encoding(i, d_model)}")
```
