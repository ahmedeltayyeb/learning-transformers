```python
import numpy as np
import scipy
```

### Generate 4 random tokens

```python
token_1 = np.random.rand(8)
token_2 = np.random.rand(8)
token_3 = np.random.rand(8)
token_4 = np.random.rand(8)
```

```python
tokens = np.array([token_1, token_2, token_3, token_4])
```

### Generate Query, Key and Value matrices

```python
w_q = np.random.rand(8, 128)
w_k = np.random.rand(8, 128)
w_v = np.random.rand(8, 128)
```

### Calculate q, k and v values, using matrix multiplication

```python
q = tokens @ w_q
```

```python
k = tokens @ w_k
```

```python
v = tokens @ w_v
```

### Attention pattern

```python
q_k = (q @ k.T) / np.sqrt(128)
```

### Apply softmax

```python
normalized_q_k = scipy.special.softmax(q_k, axis=-1)
```

### Attention output

```python
attention_output = normalized_q_k @ v
```

```python
normalized_q_k.shape
```

### Final projection

```python
W_v_up = np.random.rand(128, 8)
```

```python
final_output = attention_output @ W_v_up
```

```python
final_output.shape
```

### Final embeddings

```python
final_embeddings = final_output + tokens
```
